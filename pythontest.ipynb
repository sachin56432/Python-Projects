```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(r"C:\Users\Sachin\Downloads\sales_data.csv")

df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])


df["Month"] = df["Sale_Date"].dt.to_period("M")


top2_categories = (
    df.groupby("Product_Category")["Sales_Amount"]
      .sum()
      .nlargest(2)
      .index
)


monthly_sales = (
    df[df["Product_Category"].isin(top2_categories)]
    .groupby(["Month", "Product_Category"])["Sales_Amount"]
    .sum()
    .unstack()  
)


monthly_sales.plot(marker="o")

plt.xlabel("Month")
plt.ylabel("Total Sales Amount")
plt.title("Monthly Sales by Top-2 Product Categories")
plt.legend(loc="lower right")
plt.grid(True, linestyle="--")

plt.show()
```
